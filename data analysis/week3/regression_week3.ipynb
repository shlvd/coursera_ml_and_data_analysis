```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('botswana.tsv', sep='\t')
df.head()
```

```python
df.religion.nunique()
```

```python
df.dropna().shape[0]
```

```python
df['nevermarr'] = list(map(lambda x: 1 if np.isnan(x) else 0, df['agefm']))
df.drop(['evermarr'], axis=1, inplace=True)
df.head()
```

```python
df['agefm'].fillna(0.0, inplace=True)
for i, item in enumerate(df['nevermarr']):
    if item == 1:
        df.heduc[i] = -1.0
df.head()
```

```python
df.heduc.isna().sum()
```

```python
df['idlnchld_noans'] = list(map(lambda x: 1 if np.isnan(x) else 0, df['idlnchld']))
df['heduc_noans'] = list(map(lambda x: 1 if np.isnan(x) else 0, df['heduc']))
df['usemeth_noans'] = list(map(lambda x: 1 if np.isnan(x) else 0, df['usemeth']))
df.head()
```

```python
for i, item in enumerate(df['idlnchld_noans']):
    if item == 1:
        df.idlnchld[i] = -1.0
for i, item in enumerate(df['heduc_noans']):
    if item == 1:
        df.heduc[i] = -2.0
for i, item in enumerate(df['usemeth_noans']):
    if item == 1:
        df.usemeth[i] = -1.0
df.dropna(inplace=True)
df.shape[0]*df.shape[1]
```

```python
import statsmodels.formula.api as smf

model = smf.ols('ceb ~ age + educ + religion + idlnchld + knowmeth + usemeth + agefm + heduc +'\
             'urban + electric + radio + tv + bicycle + nevermarr + idlnchld_noans +'\
             'heduc_noans + usemeth_noans', 
             data=df)
fitted = model.fit()
print(fitted.summary())
```

```python
import statsmodels.stats.api as sms
print('Breusch-Pagan test: p={}'.format(sms.het_breushpagan(fitted.resid, fitted.model.exog)[1]))
```

```python
df.drop(['religion', 'radio', 'tv'], axis=1, inplace=True)
model1 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + usemeth + agefm + heduc +'\
             'urban + electric + bicycle + nevermarr + idlnchld_noans +'\
             'heduc_noans + usemeth_noans', 
             data=df)
fitted1 = model1.fit(cov_type='HC1')
print(fitted1.summary())
```

```python
print('Breusch-Pagan test: p={}'.format(sms.het_breushpagan(fitted1.resid, fitted1.model.exog)[1]))
```

```python
print(model.fit().compare_f_test(model1.fit()))
```

```python
model2 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + agefm + heduc +'\
             'urban + electric + bicycle + nevermarr + idlnchld_noans +'\
             'heduc_noans', 
             data=df)
fitted2 = model2.fit(cov_type='HC1')
print(fitted2.summary())
```

```python
print(model1.fit().compare_f_test(model2.fit()))
```
