```python
# import all necessary modules
%matplotlib inline

import numpy as np
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt
```

## For this assignment i choosed the chi-squared distribution (also chi-square or Ï‡2-distribution) with k = 20 degrees of freedom


The variable obtained by summing the squares of df independent, standard normally distributed random variables:

$Q = \sum_{i=0}^{\mathtt{df}} X^2_i$


is chi-square distributed, denoted

$Q \sim \chi^2_k.$


The probability density function of the chi-squared distribution is

$p(x) = \frac{(1/2)^{k/2}}{\Gamma(k/2)} x^{k/2 - 1} e^{-x/2},$

where \Gamma is the gamma function,

$\Gamma(x) = \int_0^{-\infty} t^{x - 1} e^{-t} dt.$

```python
#hist of generated samples
x = chi2.rvs(20, size=1000)
plt.hist(x, 40, normed=True, label='Sample')

# pdf of chi2 with k = 20
x1 = np.arange(0,50,0.5)
plt.plot(x1, chi2.pdf(x1, df=20),'r',lw=2, label='Chi2 pdf')
plt.legend()
plt.xlabel('x')
plt.ylabel('Pdf')
```

Sample distribution hist is similar to chi2 pdf as expected

```python
# sample hist+ norm pdf for different sample size 
chi_2 = chi2(20)
for n in [5, 10, 50]:
    # mean from each sample
    sample_mean = []
    for i in range(1000):
        sample_mean.append(chi_2.rvs(n).mean())
    plt.hist(sample_mean, 40, normed=True, label='Sample means')
    # for chi2 mean = k and variance = 2*k, where k(degrees of freedom) in our case is 20
    norm_dist = norm(20, np.sqrt(2*20/n))
    x = np.linspace(5, 40, 100)
    pdf = norm_dist.pdf(x)
    plt.plot(x, pdf, 'r', label='Norm dist pdf')
    plt.legend()
    plt.title("Sample size %d" % n)
    plt.ylabel('$f(x)$')
    plt.xlabel(' $x$')
    plt.show()
```

The Central Limit Theorem is exactly what the shape of the distribution of means will be when we draw repeated samples from a given population. Specifically, as the sample sizes get larger, the distribution of means calculated from repeated sampling will approach normality.
